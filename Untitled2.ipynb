```python
import numpy as np
from scipy import signal
from scipy.fft import fftshift
from matplotlib import mlab
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
import numpy.random as rng
import scipy
import scipy.interpolate as interpolate

def log_inter(
              spectrogram,
              freq, time,
              out_size):


    out_f, out_t = out_size

    log_f = np.log(freq + freq[1])

    log_f_normalized = (log_f - log_f[0]) / (log_f[-1] - log_f[0])
    f = out_f * log_f_normalized

    t_normalized = (time - time[0]) / (time[-1] - time[0])
    t = out_t * t_normalized

    out_spectrogram = np.zeros((out_f, out_t), dtype=np.float64)

    f_i = np.arange(out_f)
    t_i = np.arange(out_t)


    spectrogram_fn = interpolate.interp2d(t, f, spectrogram, copy=False)
    out_spectrogram[:, :] = spectrogram_fn(f_i, t_i)

    f_fn = interpolate.interp1d(f, freq, copy=False)
    t_fn = scipy.interpolate.interp1d(t, time, copy=False)

    f_interpolated = f_fn(f_i)
    t_interpolated = t_fn(t_i)

    return f_interpolated, t_interpolated, out_spectrogram

fig, r = plt.subplots(4, 2)
fig.set_figheight(15)
fig.set_figwidth(15)

for i,f_ in enumerate([0.5,1,2,3]):
    mean_f = f_ * 1e3
    fs = 10e3
    N = 1e6
    NFFT = 1024
    amp = 2 * np.sqrt(2)
    noise_power = 0.01 * fs / 2
    time = np.arange(N) / float(fs)
    mod = 500*np.cos(2*np.pi*0.25*time)
    carrier = amp * np.sin(2*np.pi*mean_f*time)

    x = carrier

    f, t, Sxx = signal.spectrogram(x, fs)
    r[i, 0].pcolormesh(t, f, Sxx, shading='gouraud')

    f_i, t_i, Sxx_i = log_inter(Sxx,f,t,(48,48))
    r[i, 1].pcolormesh(Sxx_i, shading = 'gouraud')
    
plt.show()


```
